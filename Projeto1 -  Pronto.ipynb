# Ánalise 3

A partir dos três gráficos acima é possível ver de forma mais específica as condições, específicamente a renda domiciliar per capta, dos domicílios pertencentes a área rural que não possuem acesso a internet. A mudança mais notável dos gráficos é a diminuição da menor renda sem que ocorra uma distribuição desses valores para as outras rendas, de forma que é possível deduzir que essa população adquiriu acesso a internet entre os anos de 2015 e 2013.

```python
ct=(pd.crosstab(dados_2013.AguaCanalizada, dados_2013.AcessoaInternetnoDomicílio, margins=True, rownames=['Água Canalizada'], colnames=['Acesso à Internet'], normalize='index')*100).round(decimals=0)


plot = ct.plot(kind='bar', stacked=True,title='Pessoas com Água Canalizada  possuem acesso à internet?')
plt.ylabel('Freq.Relativa (em %)')

#Para legenda ficar fora do gráfico
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

print('Segundo o gráfico, entre as pessoas que possuem água encanada, cerca de 20% delas possuem acesso a internet, apenas.')
print('Entre as que não possuem água encanada, há uma ocorrencia de cerca de 5% dessas pessoas possuirem acesso a internet.')
print('Num geral, boa parte da amostra não está concetada a rede.')
```

# Conclusão

Com o projeto é possível concluir que existe no Brasil rural uma situação complexa. Há uma pequena parte da população que não
possui água canalisada mas possui acesso a internet em seu domicilio. Desse fato é possivel fazer diversas interpretações, como:
